```r
# build the function top_25

library(stringr)
top_25 <- function(team_name,year){
    #https://www.spotrac.com/mlb/atlanta-braves/payroll/   :2019
    #https://www.spotrac.com/mlb/atlanta-braves/payroll/2017/   
    Tname <- str_replace_all(tolower(team_name), pattern = " ",replacement = "-")
    Year <- ifelse(year == 2019,"",as.character(year))
    
    # be careful about the Tname any excess space will lead to wrong!!
    url <- paste("https://www.spotrac.com/mlb/",Tname,"/payroll/",Year,"/",sep="")  #remember the 'sep' argument
    rd <- read_html(url)
    node1 <- html_nodes(rd,css=".teams")
    node2 <- html_node(node1,css=".datatable")
    table <- html_table(node2)
    return (table)
}
```

```r
top_25("baltimore-orioles",2017)
```

```r

# get team name first
rd <- read_html("http://m.mlb.com/teams/")
node_team <- html_nodes(rd,"h3")
team_name <- html_text(node_team)
print(team_name)

# get the year 
year <- 2011:2019
print(year)
```

```r
# get all top 25 man's salary each year and each team

for (i in team_name){
    for (j in year){
        table <- top_25(i,j)
        #output the csv file 
        file_name <- paste(i," ",j,".csv",sep="")
        write.table(table,file_name,sep=",",row.names = FALSE)
        
    }
}
```
